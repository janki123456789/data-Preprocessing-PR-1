### Que 1

Data Analysis is the process of collecting, cleaning, organizing, and interpreting data to extract useful information and support decision-making.

It helps businesses and organizations understand patterns, trends, and insights from raw data.

Purpose:

- Identify patterns, trends, and relationships within the data.
- Detect anomalies or outliers that may impact results.

Example:

A company analyzes sales data to find:
- Which product sells the most
- Which season has higher sales

### Que 2


A Data Science project requires structured planning. The main steps are:

1. Problem Understanding

- Define the problem clearly
- Understand business objectives

2. Data Collection

- Gather data from sources like CSV, databases, APIs

3. Data Cleaning

- Handle missing values
- Remove duplicates

4. Exploratory Data Analysis (EDA)

- Analyze patterns and relationships
- Use charts and graphs

5. Feature Engineering

- Select important variables
- Create new features if needed

6. Model Building

- Apply Machine Learning algorithms

7. Model Evaluation

- Check performance using metrics (Accuracy, Precision, Recall)

8. Deployment

- Deploy the model in real-world applications

9. Monitoring & Maintenance

- Continuously monitor performance
- Update the model when needed

### Que 3

Problem Statement:

The goal is to predict whether a customer will stop using a company’s service (churn) based on their purchase behavior.

Objective:

To build a model that identifies customers likely to churn.

Input Features:

- Purchase frequency
- Total spending
- Last purchase date
- Payment method
- Customer demographics (age, location)

Output:

0 → Customer will not churn
1 → Customer will churn

Type of Problem:

Supervised Learning (Classification)

Example:

- If a customer hasn’t made a purchase for a long time, they are more likely to churn.

### Que 4

A Tensor is a mathematical object that generalizes scalars (0D), vectors (1D), and matrices (2D) to higher dimensions. In simple terms, a tensor is just a container for data in multiple dimensions, widely used in deep learning and numerical computing.


| Type         | Example       | Description             |
| ------------ | ------------- | ----------------------- |
| Scalar (0D)  | 5             | Single value            |
| Vector (1D)  | [1, 2, 3]     | One-dimensional array   |
| Matrix (2D)  | [[1,2],[3,4]] | Two-dimensional array   |
| Tensor (3D+) | [[[1]]]       | Multi-dimensional array |


```python
import numpy as np

a = np.array(10)

b = np.array([1, 2, 3])

c = np.array([[1, 2], [3, 4]])

d = np.array([[[1, 2], [3, 4]],
              [[5, 6], [7, 8]]])

print(a)
print(b)
print(c)
print(d)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer_dataset.csv')

sns.histplot(df['Income'], kde=True,  bins=20, color='blue', edgecolor='white', alpha=0.85)
plt.title("Income Distribution")
plt.show()

```

```python
# ─── KDE Plot  ───
import matplotlib.pyplot as plt

sns.kdeplot(df['Income'], color='blue', fill=True, alpha=0.4)
plt.title("Income Distribution")
plt.show()
```

```python
# ─── Box Plot  ───
import matplotlib.pyplot as plt

plt.boxplot(df['Purchases'], patch_artist=True )
plt.title("Purchase Distribution")
plt.show()
```
